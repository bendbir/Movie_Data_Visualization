```python
# Import plotly offline
import plotly
plotly.offline.init_notebook_mode(connected=True)
import plotly.graph_objs as go
import numpy as np
import pandas as pd

# Import pandas
import pandas as pd
```

```python
movies = pd.read_csv('movies_metadata.csv', low_memory=False)
```

```python
print (movies.head())
```

```python
# Much more data than we need.
# Select just original title, years, budget, revenue, and length
edit_movies_df = movies.drop(['adult',
                              'belongs_to_collection',
                              'genres','homepage',
                              'id','imdb_id',
                              'original_language',
                              'original_title',
                              'overview','popularity',
                              'poster_path','production_companies',
                              'production_countries','spoken_languages',
                              'status','tagline','video','vote_average',
                              'vote_count'], axis=1)
```

```python
edit_movies_df.head()
```

```python
# Remove all non numeric containing rows.
edit_movies_df = edit_movies_df[pd.to_numeric(edit_movies_df['budget'], errors='coerce').notnull()]
edit_movies_df = edit_movies_df[pd.to_numeric(edit_movies_df['revenue'], errors='coerce').notnull()]
edit_movies_df = edit_movies_df[pd.to_numeric(edit_movies_df['runtime'], errors='coerce').notnull()]
```

```python
# Convert all numeric columns to floats, in case they are strings.
edit_movies_df['budget'] = edit_movies_df['budget'].astype('float64')
edit_movies_df['revenue'] = edit_movies_df['revenue'].astype('float64')
edit_movies_df['runtime'] = edit_movies_df['runtime'].astype('float64')
edit_movies_df.head()
```

```python
# Drop all NaN or 0 containing rows
edit_movies_df = edit_movies_df.replace(0, np.nan).replace(0.0, np.nan).dropna()
```

```python
edit_movies_df['release_date'] = edit_movies_df['release_date'].apply(lambda date: pd.to_datetime(date))
```

```python
# How many movies left in our list?
print (len(edit_movies_df))
edit_movies_df.head()
```

```python
# Group movies by year, this is how we'll be sorting them later
date_sort = edit_movies_df.sort_values('release_date')
print (date_sort.head())
```

```python
# Assign dates to list
date_list = date_sort['release_date'].tolist()

# Assign revenue to list
revenue_list = date_sort['revenue'].tolist()

# Assign budget to list
budget_list = date_sort['budget'].tolist()

# Assign runtime to list
runtime_list = date_sort['runtime'].tolist()
```

```python
# Graph using plotly, after generating series of x and y axis
trace1 = go.Bar(x = date_list, y = budget_list, name = 'Budget')

data = [trace1]

layout = dict(title = 'Movie Budget Over Time',
              xaxis = dict(title = 'Date'),
              yaxis = dict(title = 'Dollars'),
              )

fig = dict(data=data, layout=layout)
plotly.offline.iplot(fig, filename='Movie Budget Over Time')
```

```python
# Graph using plotly, after generating series of x and y axis
trace2 = go.Bar(x = date_list, y = revenue_list, name = 'Revenue')

data = [trace2]

layout = dict(title = 'Movie Revenue Over Time',
              xaxis = dict(title = 'Date'),
              yaxis = dict(title = 'Dollars'),
              )

fig = dict(data=data, layout=layout)
plotly.offline.iplot(fig, filename='Movie Revenue Over Time')
```

```python
# Graph using plotly, after generating series of x and y axis
trace3 = go.Bar(x = date_list, y = runtime_list, name = 'Runtime')

data = [trace3]

layout = dict(title = 'Movie Runtime Over Time',
              xaxis = dict(title = 'Date'),
              yaxis = dict(title = 'Minutes'),
              )

fig = dict(data=data, layout=layout)
plotly.offline.iplot(fig, filename='Movie Runtime Over Time')
```

```python
# This data is interesting, but mean for each year would be better
date_sort['YEAR'] = date_sort['release_date'].apply(lambda x: x.strftime('%Y'))
```

```python
date_sort.head()
```

```python
# EXAMINING SPECIFIC YEARS

# Let's test what movies were released during specific years,
# To see what causes the spikes we see on the average graph.
specific_year = date_sort.loc[date_sort['YEAR'] == '2017']
specific_year
```

```python
# Group by years, find means
years_mean_df = date_sort.groupby('YEAR', as_index=True).mean()
#years_mean_df = years_mean_df.mean().sort_values('YEAR')
```

```python
years_mean_df.head()
```

```python
# Generate YEAR_list, from original date_sort
YEAR_list = date_sort['YEAR'].drop_duplicates().tolist()
type(YEAR_list)
print (YEAR_list[0:10])
```

```python
# Create new variable lists for x and y axis
# Generate YEAR_list, from original date_sort
YEAR_list = date_sort['YEAR'].drop_duplicates().tolist()

# Assign revenue to list
revenue_list_2 = years_mean_df['revenue'].tolist()

# Assign budget to list
budget_list_2 = years_mean_df['budget'].tolist()

# Assign runtime to list
runtime_list_2 = years_mean_df['runtime'].tolist()
```

```python
print (YEAR_list[0:10])
print (revenue_list_2[0:10])
print (budget_list_2[0:10])
print (runtime_list_2[0:10])
```

```python
# Graph using plotly, after generating series of x and y axis
trace1 = go.Scatter(x = YEAR_list, y = revenue_list_2, name = 'Revenue')
trace2 = go.Scatter(x = YEAR_list, y = budget_list_2, name = 'Budget')
data = [trace1, trace2]

layout = dict(title = 'Average Movie Revenue and Budget By Year',
              xaxis = dict(title = 'Year'),
              yaxis = dict(title = 'Dollars'),
              )

fig = dict(data=data, layout=layout)
plotly.offline.iplot(fig, filename='Movie Budget Over Time')
```

Revenue represents total lifetime gross, not just box office sales while in theatres.  This may explain the spikes with early years.  These years contain big box office movies that have also seen re-release and sucess on home video or DVD, and once we removed rows with zeroes some years have fewer movies.  1937 is only represented by "Snow White and the Seven Dwarfs", which has enjoyed massive success in re-release and home video.  1939 was when "Gone with the Wind" and "The Wizard of Oz" were released, 1942 was "Bambi" and "Casablanca".  A later spike, 1977, was when "Star Wars" came out, another film that's had re-release in theatres and massive success on VHS and DVD.

1927's revenue mean seems to be skewed by the reported budget of "Metropolis" in this dataset, ~92 million US dollars.  The budget I found reported online was 5.3 million Reichsmarks, which converts to roughly 18 million US dollars today.  I'm not sure where the source of this dataset got their budget.

Looking at the raw graphs, another outlier is in 1962.  This corresponds with "The Manchurian Candidate".  Again, this dataset reports a higher budget than what is found online through other sources.  This dataset reports 80 million US dollars, when the budget is reported as 2.2 million online.  Even when converted to adjust for inflation, this only equals 18 million US dollars.

```python
# Average runtime each year
# Graph using plotly, after generating series of x and y axis
trace3 = go.Scatter(x = YEAR_list, y = runtime_list_2, name = 'Runtime')

data = [trace3]

layout = dict(title = 'Average Movie Runtime By Year',
              xaxis = dict(title = 'Year'),
              yaxis = dict(title = 'Minutes'),
              )

fig = dict(data=data, layout=layout)
plotly.offline.iplot(fig, filename='Movie Runtime Over Time')
```

Why are the runtime means so much more variable?  After removing rows containing 0 or NaNs, many early years are only represented by one movie.  This causes radical variability in mean.  Once we reach the 1970s there are films with more complete information.  Then we see a fairly stable mean of ~110 minutes from the 1980s up through the present.
